```python
import numpy as np

sl = np.load("saliencymap_0.8-Necessity.npy")
ig = np.load("integratedgrad_0.8-Necessity.npy")
sm = np.load("smoothgrad_0.8-Necessity.npy")
gradcam = np.load("gradcam_0.8-Necessity.npy")
neuron = np.load("neuroninfl_0.8-Necessity.npy")
channel = np.load("channelinfl_0.8-Necessity.npy")
```

```python
print("Saliency Map: ", sl.mean())
print("Integrated Grad: ", ig.mean())
print("Smooth Grad: ", sm.mean())
print("GradCAM: ", gradcam.mean())
print("neuron: ", neuron.mean())
print("channel: ", channel.mean())
```

```python
import matplotlib.pyplot as plt 

plt.plot(np.arange(len(gradcam)), gradcam)
plt.show()
```

```python
labels = ['Saliency Map', 'IG', 'SM', 'GradCAM']
x_pos = np.arange(len(labels))
CTEs =[sl.mean(), ig.mean(), sm.mean(), gradcam.mean()]
error = [sl.std(), ig.std(), sm.std(), gradcam.std()]
fig, ax = plt.subplots()
ax.bar(x_pos, CTEs,
       yerr=error,
       align='center',
       alpha=0.5,
       ecolor='black',
       capsize=10)
ax.set_ylabel('0.8-Necessity')
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_title('Attribution Methods')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('bar_plot_with_error_bars.png')
plt.show()
```
